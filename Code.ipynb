```python
# /Users/vmatsiiako/Downloads/X1-3.jld
import h5py
import numpy as np

jld_fn = "Optimization-Project/X1-3.jld"
with h5py.File(jld_fn, "r") as f:
    a_group_key = list(f.keys())[0]
    X = np.array(f[a_group_key]).T
```

```python
import pandas as pd
from scipy.spatial import distance_matrix
D = pd.DataFrame(distance_matrix(X, X))
```

```python
from itertools import product
from math import sqrt

import gurobipy as gp
from gurobipy import GRB
```

```python
m = gp.Model('depo_assignment')

num_clients = len(X)
d_max = 10
n_min = 2
n_max = 8
# num_depos = m.addVars([1], vtype=GRB.INTEGER, name='num_depos')
num_depos = 4
cartesian_prod_X = list(product(range(num_clients), range(num_depos)))
# print(cartesian_prod_X)
assignment = m.addVars(cartesian_prod_X, vtype=GRB.BINARY, name='assignment')
cartesian_prod = list(product(range(num_clients), range(num_clients), range(num_depos)))

# m.addConstrs((assign[(c,f)] <= select[f] for c,f in cartesian_prod), name='Setup2ship')
# m.addConstrs((gp.quicksum(assign[(c,f)] for f in range(num_facilities)) == 1 for c in range(num_customers)), name='Demand')
m.addConstrs((gp.quicksum(assignment[(c,f)] for f in range(num_depos)) <= 1 for c in range(num_clients)), name='Unique')
m.addConstrs((gp.quicksum(assignment[(f,c)] for f in range(num_clients)) >= n_min for c in range(num_depos)), name='Min')
m.addConstrs((gp.quicksum(assignment[(f,c)] for f in range(num_clients)) <= n_max for c in range(num_depos)), name='Max')
# print(D[0][0])
m.addConstrs((D[i][k] * assignment[(i,j)] * assignment[(k,j)] <= d_max for i,k,j in cartesian_prod), name='Setup2ship')



# m.setObjective(num_depos, GRB.MINIMIZE)
obj = gp.quicksum(assignment[(c,f)] for c,f in cartesian_prod_X)
m.setObjective(obj, GRB.MAXIMIZE)
#                gp.quicksum(assignment[(c,f)] for f in range(num_depos) for c in range(num_clients))

m.optimize()
```
